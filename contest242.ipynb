```python
class Solution:
    def minSpeedOnTime(self, dist: List[int], hour: float) -> int:
        n=len(dist)
        if n >= hour+1:
            return -1
        sum1=sum(dist)
        v=math.ceil(sum1/hour)
        start=v
        if hour%1==0:
            end=max(dist)
        else:
            end=max(dist)/(hour%1)
        
        while start<end:
            v=(start+end)//2
            num=0
            for i in range((len(dist))-1):
                num+=math.ceil(dist[i]/v)
            else:
                i+=1
                num+=dist[i]/v
            if num<=hour:
                end=v
            else:
                start=v+1

        return start
        # dist.sort(reverse=True)
        # time = floor(hour+1)-n
        # if time==0:
        #     return max(dist[-1]/(hour-floor(hour)),dist[0])
        # else:

```
